```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from datetime import datetime
from datetime import timedelta
from time import time
```

```python
## Data Source: scraped from histdata.com
### https://drive.google.com/drive/folders/1QYIsRmIKQ7Jbxsla82CLyx5RDwJEdRyF?usp=drive_link
```

```python
df = pd.read_csv('DAT_ASCII_WTIUSD_M1_2011.csv', sep = ';', header = None, parse_dates = [0])
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'x']
df
```

```python
for year in np.arange(2011, 2024, 1):
    filename = f'DAT_ASCII_WTIUSD_M1_{year}.csv'
    newdata = pd.read_csv(filename, sep = ';', header = None, parse_dates = [0])
    newdata.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'x']
    df = pd.concat([df, newdata])
    
df.drop(['High', 'Low', 'x'], axis = 1, inplace = True)
    
```

```python
df.to_csv('WTI_Minute_Data.csv')
```

```python
df[(df.Date > '2011-01-01') & (df.Date < '2018-01-01')].to_csv('WTI_Minute_Data_2011_2017.csv')
df[(df.Date > '2018-01-01')].to_csv('WTI_Minute_Data_2018_2023.csv')
```
